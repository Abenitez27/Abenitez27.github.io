## This is a simple web scraper for scraping one page of data at a time. This process can be extended to multiple pages but requires a more in depth nested loop with prescraping of the webpage and multiple rotating proxies. The user is required to go to Zillow with thier preferred search engine I used Google Chrome. Then search for a city and copy the URL and paste it into the input prompt.

```python
zillow_url = input("Please enter your Zillow URL here: ")
```

```python
# Imports:
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen as uReq
import pandas as pd
import re

```

```python
#Header is used as a proxy for fooling Zillow into thinking we're a user browser:
headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'}
```

```python
#URL request command for collecting HTML:
r = requests.get(zillow_url, headers=headers)
c = r.content
soup = BeautifulSoup(c, "html.parser")

table = []

# Grabs all photo cards from web page and assigns it to variable photo_cards     
photo_cards = soup.findAll("article",{"class":"list-card list-card_not-saved"})

# scrapes data for each photo card and loops through entire webpage
for card in photo_cards:
    try:
        address = card.div.address.text.replace(",","|")
        try:
            street = address.split("|")[0]
            city = address.split("|")[1]
            state = address.split("|")[2].split(" ")[1]
            zip_code= address.split("|")[2].split(" ")[2]
        except:
            pass 
    except:
        pass
    try:
        list_price = card.find("div",{"class": "list-card-price"}).text.replace(",","").replace("$","")
    except:
        pass
    try:
        num_beds = card.findAll("li")[1].text
    except:
        pass
    try:
        num_baths = card.findAll("li")[2].text
    except:
        pass
    try:
        num_sqft = card.findAll("li")[3].text.replace(",","")
    except:
        pass
    try:
        misc_info = card.find("div",{"class":"list-card-variable-text list-card-img-overlay"}).text
    except:
        pass
    try:
        sale_type = card.find("div", {"class":"list-card-type"}).text
    except:
        pass

    # appends data to the list vairable "table"
    try:
        table.append([street, city, state, zip_code, list_price, num_beds, num_baths, num_sqft, misc_info, sale_type])
    except:
        pass

        
# creates headers, dataframe and shows dataframe head        
headers = ["street","city","state","zip_code", "list_price", "num_beds", "num_baths", "num_sqft", "misc_info", "sale_type"]
df=pd.DataFrame(data= table, columns=headers)
df.head()
```
